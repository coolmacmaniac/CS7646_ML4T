```python
import json
import yaml
```

```python
data_dict1 = [
    {
        "inputs": {
            "start_date": "2010-01-01",
            "end_date": "2010-12-31",
            "symbols": [
                "GOOG",
                "AAPL",
                "GLD",
                "XOM"
            ],
        },
        "outputs": {
            "allocs": [
                0.0,
                0.4,
                0.6,
                0.0
            ]
        },
        "description": "Example 1"
    },
    {
        "inputs": {
            "start_date": "2004-01-01",
            "end_date": "2006-01-01",
            "symbols": [
                "AXP",
                "HPQ",
                "IBM",
                "HNZ"
            ],
        },
        "outputs": {
            "allocs": [
                0.78,
                0.22,
                0.0,
                0.0
            ]
        },
        "description": "Example 2"
    }
]
```

```python
print(data_dict1)
```

```python
print(json.dumps(data_dict1, indent=4))
```

```python
with open('test_data1.yaml', 'w') as fp:
    yaml.dump(data_dict1, fp)
```

```python
with open('test_data1.yaml', 'r') as fp:
    data = yaml.load(fp, Loader=yaml.SafeLoader)
    print(json.dumps(data, indent=4))
```
